```python
datas = []
temp = [[] for i in range(len(keys))]
for i in data['data'].keys():
    temp_data = data['data'][i]
    
    for j in range(len(keys)):
        # if keys[j] in ['references', 'proteins']:
        #     continue
        try:
            if keys[j] == 'ph_range':
                in_temp = []
                for k in range(len(temp_data[keys[j]])):
                    ph = str(temp_data[keys[j]][k]['min_value']) + '-' + str(temp_data[keys[j]][k]['max_value'])
                    in_temp.append(ph)
                temp[j].append(in_temp)
            elif keys[j] == 'km_value':
                in_temp = []
                for k in range(len(temp_data[keys[j]])):
                    km = str(temp_data[keys[j]][k]['num_value'])
                    in_temp.append(km)
                temp[j].append(in_temp)
            elif keys[j] == 'inhibitor':
                in_temp = []
                for k in range(len(temp_data[keys[j]])):
                    inhibitor = str(temp_data[keys[j]][k]['value']) + ':' + str(temp_data[keys[j]][k]['comment'])
                    in_temp.append(inhibitor)
                temp[j].append(in_temp)
            elif keys[j] == 'specific_activity':
                in_temp = []
                for k in range(len(temp_data[keys[j]])):
                    in_temp.append(str(temp_data[keys[j]][k]['num_value']))
                temp[j].append(in_temp)
            elif keys[j] == 'synonyms':
                in_temp = []
                for k in range(len(temp_data[keys[j]])):
                    in_temp.append(str(temp_data[keys[j]][k]['value']))
                temp[j].append(in_temp)
            elif keys[j] == 'cofactor':
                in_temp = []
                for k in range(len(temp_data[keys[j]])):
                    in_temp.append(str(temp_data[keys[j]][k]['value']))
                temp[j].append(in_temp)
            elif keys[j] == 'subunits':
                in_temp = []
                for k in range(len(temp_data[keys[j]])):
                    in_temp.append(str(temp_data[keys[j]][k]['value']))
                temp[j].append(in_temp)
            elif keys[j] == 'reaction_type':
                in_temp = []
                for k in range(len(temp_data[keys[j]])):
                    in_temp.append(str(temp_data[keys[j]][k]['value']))
                temp[j].append(in_temp)
            elif keys[j] == 'purification':
                in_temp = []
                for k in range(len(temp_data[keys[j]])):
                    in_temp.append('organisms' + ':' + str(temp_data[keys[j]][k]['organisms'][0]))
                temp[j].append(in_temp)
            elif keys[j] == 'general_stability':
                in_temp = []
                for k in range(len(temp_data[keys[j]])):
                    in_temp.append('organisms' + ':' + str(temp_data[keys[j]][k]['organisms'][0]))
                temp[j].append(in_temp)
            elif keys[j] == 'storage_stability':
                in_temp = []
                for k in range(len(temp_data[keys[j]])):
                    in_temp.append('organisms' + ':' + str(temp_data[keys[j]][k]['organisms'][0]))
                temp[j].append(in_temp)
                
            elif keys[j] in ['natural_reaction', 'generic_reaction', 'reaction']:
                in_temp = []
                for k in range(len(temp_data[keys[j]])):
                    in_temp.append('educts:' + str(temp_data[keys[j]][k]['educts']) + ',' + 'products:' + str(temp_data[keys[j]][k]['products']))
                temp[j].append(in_temp)
            elif keys[j] in ['molecular_weight']:
                in_temp = []
                for k in range(len(temp_data[keys[j]])):
                    in_temp.append(str(temp_data[keys[j]][k]['num_value']))
                temp[j].append(in_temp)
            else:
                temp[j].append(temp_data[keys[j]])
        except:
            temp[j].append('none')
            
import pandas as pd
pd_data = pd.DataFrame()
for i in range(len(keys)):
    # if keys[i] in ['references', 'proteins']:
    #     continue
    pd_data[keys[i]] = temp[i]
```
